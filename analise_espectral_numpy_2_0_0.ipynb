```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Função para carregar o sinal de um arquivo
def load_signal(file_path):
    tempo_acelera = []
    aceleracao = []

    with open(file_path) as acelera:
        linhas_acelera = acelera.read().split('\n')
        for linha in linhas_acelera:
            if linha.strip():  # Verificar se a linha não está vazia
                elementos = linha.split('\t')
                tempo_acelera.append(float(elementos[0]))
                aceleracao.append(float(elementos[1]))

    return np.array(tempo_acelera), np.array(aceleracao)
```

```python
# Função para realizar a análise espectral
def analyze_spectrum(signal, sampling_rate):
    # Número de amostras no sinal
    N = len(signal)

    # Aplicar a FFT
    fft_values = np.fft.fft(signal)

    # Frequências correspondentes
    freqs = np.fft.fftfreq(N, d=1/sampling_rate)

    # Magnitude do espectro
    magnitude = np.abs(fft_values)

    return freqs, magnitude
```

```python

# Função para plotar o sinal e seu espectro
def plot_signal_and_spectrum(tempo, sinal, freqs, magnitude):
    plt.figure(figsize=(12, 6))

    # Plotar o sinal original
    plt.subplot(2, 1, 1)
    plt.plot(tempo, sinal)
    plt.title("Sinal Original")
    plt.xlabel("Tempo (s)")
    plt.ylabel("Aceleração (m/s²)")

    # Plotar o espectro de frequência
    plt.subplot(2, 1, 2)
    plt.plot(freqs, magnitude)
    plt.title("Espectro de Frequência")
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Magnitude")
    plt.xlim(0, max(freqs) / 2)  # Plotar apenas a metade positiva do espectro

    plt.tight_layout()
    plt.show()
```

```python

# Caminho para o arquivo contendo o sinal
file_path = '/content/drive/MyDrive/Colab Notebooks/experimentos_viga/ensaio01_a.txt'


# Taxa de amostragem (em Hz)
sampling_rate = 1000  # Exemplo de taxa de amostragem

# Carregar o sinal
tempo, sinal = load_signal(file_path)

# Realizar a análise espectral
freqs, magnitude = analyze_spectrum(sinal, sampling_rate)

# Plotar os resultados
plot_signal_and_spectrum(tempo, sinal, freqs, magnitude)
```
